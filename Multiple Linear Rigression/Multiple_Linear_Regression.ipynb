```python
# Find the dependent Varoable & Independenyt Variable
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Convert the column in to the Categorical column.
```

```python
state = pd.get_dummies(X["State"], drop_first= True)
```

```python
state
```

```python
X = X.drop("State", axis=1)
```

```python
X = pd.concat([X,state], axis = 1)
```

```python
X.head()
```

```python
# Splitting the dataset in train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_test,X_train,y_test,y_train = train_test_split(X,y, test_size= 0.2, random_state= 0)
```

```python
# Fitting multiple linear regression in to the Training dataset
```

```python
from sklearn.linear_model import LinearRegression
lin_regressor = LinearRegression()
lin_regressor.fit(X_train,y_train)
```

```python
y_predict = lin_regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_predict)
```
