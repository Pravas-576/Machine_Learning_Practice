```python
from sklearn import datasets
cancer = datasets.load_breast_cancer()
cancer
```

```python
print("Features :", cancer.feature_names)
print("Labels :", cancer.target_names)
```

```python
cancer.data.shape
```

```python
print(cancer.data[0:5])
```

```python
print(cancer.target)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(cancer.data, cancer.target, test_size = 0.25, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.svm import SVC
model = SVC(kernel = 'linear')
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn import metrics
print("Accuracy Score :", metrics.accuracy_score(y_test,y_pred))
```

```python
# Model Precision: what percentage of positive tuples are labeled as such? TP / (TP + FP)
print("Precision :", metrics.precision_score(y_test,y_pred))
# Model Recall: what percentage of positive tuples are labelled as such? TP / (TP + FN)
print("Recall :", metrics.recall_score(y_test,y_pred))
```
